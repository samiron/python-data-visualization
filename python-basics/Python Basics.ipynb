# Condition
General syntax
```python
if <condition>:
    statement
elif <condition>:
    statement
else:
    statement
```

A quick shorthand

```python
statement if condition else statement
```

# `Pass`
Python provides a special keyword `pass` to represent an empty expression block. Example,
```python
if condition:
   print("something")
else:
   pass
```

# Loop

**Iterate an Array or List**
```
fruit_names = ['Mango', 'Cherry', 'Berry', 'Avocado']
for fruit_name in fruit_names:
  print("Fruit Name is: {}".format(fruit_name));
```

**Iterate Dictionary**
```
cn_state_capitals = {'Nova Scotia': 'Halifax',
                  'Ontario': 'Toronto',
                  'Albarta': 'Edmonton'}

# Iterate using only value
for city in cn_state_capitals:
  print(city)

@Iterate using key and value
for (key, value) in cn_state_capitals.items():
  print(key, value)
```
